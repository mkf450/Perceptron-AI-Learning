```python
import numpy as np
```

```python
class Perceptron(object):
    """Implements a perceptron network"""
    def __init__(self, input_size, lr=0.2, epochs=100):
        self.W = np.zeros(input_size+1) # add one for bias
        self.epochs = epochs #epoch
        self.lr = lr #learning rate
    
    def activation_fn(self, z):
        return 1 if z >= 0 else -1
 
    def predict(self, x):
        z = np.dot(self.W,x)
        a = self.activation_fn(z)
        return a
    
    def net_acc(self,X,t):
      acc = 0
      for i in range(t.shape[0]):
        x = np.insert(X[i], 0, 1) #add bias pada indeks ke-0
        if self.predict(x) == t[i]:
          acc = acc + 1
      return acc/t.shape[0]
 
    def fit(self, X, t):
        for e in range(self.epochs):
            for i in range(t.shape[0]):
                x = np.insert(X[i], 0, 1) #add bias pada indeks ke-0
                y = self.predict(x)
                self.W = self.W + self.lr * (t[i] - y) * x
                #print(self.W)
            print("Epoch ke- ", e, "Accuracy = ", self.net_acc(X,t))
```

```python
 X = np.array([
        [1, 1],
        [1, -1],
        [-1, 1],
        [-1, -1]
    ])
t = np.array([1, 1, 1, -1])
 
print(X.shape)
print(t.shape)
```

```python
perceptron = Perceptron(input_size=2,lr=0.2,epochs=10)
perceptron.fit(X, t)
print(perceptron.W)
```
